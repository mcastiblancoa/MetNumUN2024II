## Bernoulli Naive Bayes

El modelo Bernoulli Naive Bayes fue seleccionado por su excelente desempeño en problemas de clasificación, especialmente cuando se trabaja con datos discretos o que pueden representarse de manera binaria, como identificar la ocurrencia o no de un evento. Es un modelo simple pero eficaz, además de ser rápido en el proceso de entrenamiento, lo que lo hace ideal para manejar grandes volúmenes de datos. Por estas razones, es una opción adecuada para predecir movimientos del mercado de forma eficiente.

## Variables agregada

El **On Balance Volume (OBV)** es un indicador que analiza el flujo acumulado de volumen en relación con las variaciones de precio, basado en la premisa de que el volumen suele preceder a los movimientos del precio. Cuando el precio cierra por encima del día anterior, el volumen se agrega al OBV; si cierra por debajo, se resta; y si no hay cambios en el precio, el OBV permanece constante. Un OBV en aumento sugiere una tendencia alcista, mientras que un OBV en descenso indica una tendencia bajista. Si el precio sube pero el OBV no acompaña, podría ser una señal de que la tendencia está perdiendo impulso, anticipando una posible reversión.

Por otro lado, se seleccionó el **VWMA (Media Móvil Ponderada por Volumen)**, un indicador que da más relevancia a los periodos con mayor volumen. Este método no solo toma en cuenta el precio, sino también la fuerza del mercado basada en el volumen, proporcionando una perspectiva más precisa de las tendencias al reflejar mejor la actividad del mercado.

## Variables agregadas

On Balance Volume (OBV) es un indicador de volumen que mide el flujo de volumen acumulado en relaci√≥n con los cambios de precio. El concepto detr√°s de OBV es que los cambios en el volumen preceden los movimientos de precio, y por lo tanto, se puede utilizar para confirmar la direcci√≥n de una tendencia o detectar posibles se√±ales de reversi√≥n.

**¬øC√≥mo funciona OBV?**
- Si el precio de un activo cierra m√°s alto que el cierre anterior, el volumen de ese d√≠a se acumula al OBV.
- Si el precio cierra m√°s bajo que el cierre anterior, el volumen se descarta.
- Si el precio cierra sin cambios, el OBV permanece igual.


**Interpretaci√≥n:**

OBV creciente: Si el OBV est√° subiendo, significa que el volumen de compra est√° superando al de venta, lo que generalmente indica que la tendencia alcista es fuerte.


OBV decreciente: Si el OBV est√° cayendo, indica que el volumen de venta est√° superando al de compra, sugiriendo que la tendencia bajista es fuerte.
Diversi√≥n entre OBV y precio: Si el precio est√° subiendo, pero el OBV no lo sigue, podr√≠a ser una se√±al de que la tendencia est√° perdiendo fuerza y una reversi√≥n podr√≠a estar cerca.

**Relative Strength Index (RSI)** es un indicador t√©cnico que mide la fuerza de los movimientos de precios durante un n√∫mero espec√≠fico de per√≠odos. Se utiliza para identificar condiciones de sobrecompra o sobreventa en un mercado, lo que ayuda a los traders a tomar decisiones sobre posibles puntos de entrada o salida.

**¬øC√≥mo funciona el RSI?**
- **C√°lculo:** RSI se calcula en base a la magnitud de las ganancias recientes en comparaci√≥n con las p√©rdidas recientes. Generalmente, se usa un per√≠odo de 14 d√≠as.


**Interpretaci√≥n:**
- **RSI por encima de 70:** Indica que el activo podr√≠a estar sobrecomprado, lo que podr√≠a sugerir una correcci√≥n o una reversi√≥n a la baja.
- **RSI por debajo de 30:** Sugiere que el activo podr√≠a estar sobrevendido, lo que podr√≠a indicar un posible rebote o reversi√≥n al alza.
- **Divergencias:** Si el precio del activo est√° subiendo pero el RSI no sigue la misma direcci√≥n, puede ser una se√±al de que la tendencia alcista est√° perdiendo fuerza y podr√≠a haber una correcci√≥n o reversi√≥n.
